7. From the array [5, 15, 25, 35, 45], print all elements except the first and last.

```python
array = np.array([5,15,25,35,45])
i = 0
while i < len(array):
  if (i == 0) or (i == len(array)-1):
     i += 1
     continue
  else:
    print(array[i])
  i += 1
```

19. Square only the positive numbers in the array
[2, -3, 4, -5, 6]

```python
array = np.array([2,-3,4,-5,6])
positive_squared = np.array([i*i for i in array if i > 0])
print(positive_squared)
```
